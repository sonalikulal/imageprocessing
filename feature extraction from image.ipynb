```python
#feature extraction from images
#libraries
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from skimage.io import imread,imshow

#loading th image
image1=imread('cat.jpg')
imshow(image1)
```

```python
#converting the original image to gray
image2=imread('cat.jpg',as_gray=True)
imshow(image2)
```

```python
#analysizing both of the image
#shape of the image
print(image1.shape)
print(image2.shape)
```

```python
print(image1.size)
print(image2.size)
```

```python
#feature extraction

#i.pixel features
pixel_features1=np.reshape(image2,(1500*1000))
pixel_features1.shape,pixel_features1
```

```python
#pixel featured for the colored image
pixel_features2=np.resize(image1,(1500*1000*3))
pixel_features2.resize,pixel_features2
```
