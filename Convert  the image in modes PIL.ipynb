```python
from PIL import Image
import matplotlib.pyplot as plt
# creating image object
img = Image.open(r"nature.jpg")
plt.title('original img')
plt.imshow(img)
plt.show()

# using convert method for img1
img1 = img.convert("L")
plt.title('output 1')
plt.imshow(img1)
plt.show()
 
# using convert method for img2
img2 = img.convert("1")
plt.title('output 2')
plt.imshow(img2)
plt.show()
```
