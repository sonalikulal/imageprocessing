```python
import cv2  
import numpy as np  
from matplotlib import pyplot as plt  
img = cv2.imread(r'sonali.jpg',1)  
  
kernel = np.ones((5,5),np.float32)/25  
dst = cv2.filter2D(img,-1,kernel)  
plt.subplot(121),plt.imshow(img),plt.title('Original')  
plt.xticks([]), plt.yticks([])  
plt.subplot(122),plt.imshow(dst),plt.title('Filter2D')  
plt.xticks([]), plt.yticks([])  
plt.show()  
```

```python
#writing text on images
import numpy as np
import cv2

image = cv2.imread('sonali.jpg',cv2.IMREAD_UNCHANGED)

position = (10,50)
cv2.putText(
     image, #numpy array on which text is written
     "Gudnyt", #text
     position, #position at which writing has to start
     cv2.FONT_HERSHEY_SCRIPT_COMPLEX, #font family
     4, #font size
     (255, 255, 255), #font color
     5) #font stroke
cv2.imwrite('output.png', image)
cv2.imshow('sonali.jpg',image)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
#blending
import cv2

image = cv2.imread('C:/Users/User/Downloads/tshirt.jpg',1)

position = (300,300)
text=cv2.putText(
     image, #numpy array on which text is written
     "NIKE", #text
     position, #position at which writing has to start
     cv2.FONT_HERSHEY_COMPLEX
, #font family
     1, #font size
     (0, 0, 0), #font color
     3) #font stroke
img=cv2.addWeighted(image,0.2,text,0.9,0)
cv2.imshow('image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
